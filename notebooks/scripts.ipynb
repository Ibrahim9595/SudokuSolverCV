```python
from random import shuffle
from os import listdir
from shutil import copyfile

def createTrainTestFromData(src_dir, train_dir, test_dir, shuffle_rounds=200, test_split=.1):
    data_dirs = listdir(src_dir)

    for dir_ in data_dirs:
        c_dir = src_dir + dir_ + '/'
        data  = listdir(c_dir)
        total = len(data)
        test_count = int(total * test_split)
        
        for i in range(shuffle_rounds):
            shuffle(data)
        
        for i in range(len(data)):
            print(c_dir + data[i], end='  ')
            if i < total - test_count:
                copyfile(c_dir + data[i], train_dir + dir_ + '/' + data[i])
                print('Copied to train')
            else:
                copyfile(c_dir + data[i], test_dir + dir_ + '/' + data[i])
                print('Copied to test')
```

```python
src_dir = '../output/'
test_dir = './dataset/test/'
train_dir = './dataset/train/'
createTrainTestFromData(src_dir, train_dir, test_dir)
```
